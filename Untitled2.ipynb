```python
from pandas import read_csv, datetime
import matplotlib.pyplot as plt

def parser(x):
    return datetime.strptime('190'+x, '%Y-%m')

series = read_csv('shampoo.csv', header=0, parse_dates=[0], index_col=0, squeeze=True, date_parser=parser)

series.plot()
series.head()
```

```python
X = series.values
train, test = X[0:-12], X[-12:]
```

```python
import numpy as np

def mean_squared_error(test, pred):
    n = test.shape[0]
    return np.sum((test - pred)**2)/n

predictions = []
history = train.tolist()
for i in range(len(test)):
    predictions.append(history[-1])
    history.append(test[i])
rmse = np.sqrt(mean_squared_error(test, predictions))

print(f'RMSE: {rmse:.3f}')
plt.plot(test)
plt.plot(predictions)
plt.legend(["test","predictions"])
```

```python
import pandas as pd

def timeseries_to_supervised(data, lag=1):
    df = pd.DataFrame(data)
    columns = [df.shift(i) for i in range(1, lag+1)]
    columns.append(df)
    df = pd.concat(columns, axis=1)
    df.fillna(0, inplace=True)
    return df
```

```python
X = train
supervised = timeseries_to_supervised(X, 1)
supervised.head()
```

```python
def inverse_difference(history, yhat, interval=1):
    return yhat + history[-interval]
```

```python
differenced = difference(train, 1)
inverted = pd.Series([inverse_difference(train, differenced[i], len(train)-i) for i in range(len(differenced))])
print(pd.Series(train).head())
print(differenced.head())
print(inverted.head())
```
